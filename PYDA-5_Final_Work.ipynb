## 5. Проверка взаимосвязи между несколькими переменными с использованием t-test статистики

1. Проверим гипотезу том, что люди со средней заработной платой так же удовлетворены работой, как и люди с высокой заработной платой

```python
salary = data['salary']
def salary_class(x):
        if x =='low':
            return 1
        elif x == 'medium':
                return 2
        elif x == 'high':
            return 3
        return 0
data['salary_numder'] = data['salary'].apply(salary_class)
salary_number = data['salary_numder']

medium_salary_satisfaction = data[salary_number==2]['satisfaction_level']
high_salary_satisfaction = data[salary_number==3]['satisfaction_level']
low_salary_satisfaction = data[salary_number==1]['satisfaction_level']
len(medium_salary_satisfaction), len(high_salary_satisfaction), len(low_salary_satisfaction)
```

```python
# проверим, похожи ли коэффициенты дисперсии в выборках
medium_salary_satisfaction_1 = medium_salary_satisfaction.sample(1200)
high_salary_satisfaction_1 = high_salary_satisfaction.sample(1200)
statistics.variance(medium_salary_satisfaction_1), statistics.variance(high_salary_satisfaction_1)
```

```python
stats.ttest_ind(medium_salary_satisfaction_1, high_salary_satisfaction_1, equal_var=True)
```

pvalue = 0.14 (> 0.05) - принимаем нулевую гипотезу о том, что удовлетворенность сотрудников со средней и высокой заработной платой одинаковая. Сотрудники со средним уровнем ЗП удовлетворены работой также как сотрудники с высокой ЗП
