```python
# Формируем признаки и целевую переменную: будем прогонозировать стоимость жилья в Бостоне ('medv')
#по значениям признаков (все другие показатели)
X, y = df[['ID', 'crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad',
       'tax', 'ptratio', 'black', 'lstat',]], df['medv']
```

```python
# Разделим данные на группу обучения и проверки (70/30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Приводем данные к одному масштабу
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)
X_train.head()
```

## Линейная регрессия

```python
# Построим линейную регресиию, где признаками будут ['ID', 'crim', 'zn', 'indus', 'chas', 'nox', 'rm',
#'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat',], которые будут определять стоимость
# жилья в Бостоне ('medv')
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

reg = LinearRegression().fit(X_train, y_train)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(df[['ID', 'crim', 'zn', 'indus', 'chas', 'nox', 'rm',\
                                'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat',]])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))

y_pred = reg.predict(X_test)

print("R2: \t", r2_score(y_test, y_pred))

print('Точность предсказания модели: ', reg.score(X_test, y_test))
```
