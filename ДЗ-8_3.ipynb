```python
# Задание 1
# Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок.
# Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp
# для данного значения userId


import pandas as pd
ratings = pd.read_csv('ratings-Copy1.csv')


ratings_grouped = ratings.groupby('userId').count().reset_index()
ratings_filtered = ratings_grouped[ratings_grouped['rating']>=100].head()
ratings_filtered.head()

film_fans_user_ids = ratings_filtered['userId'].tolist()

ratings_fan = ratings[ ratings['userId'].isin(film_fans_user_ids) ]
ratings_fan

ratings_fan_life_time = ratings_fan.groupby('userId').agg(['min', 'max'])['timestamp']

ratings_fan_life_time['life_time'] = ratings_fan_life_time['max']  - ratings_fan_life_time['min'] 
#ratings_fan_life_time

all_life_time = ratings_fan_life_time['life_time'].sum()
mean_life_time = all_life_time/len(ratings_fan.groupby('userId'))
mean_life_time
```
